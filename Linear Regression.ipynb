```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('insurance.csv')
df.info()
```

```python
df
```

```python
df.describe()   # describe gives stats summary
```

```python
plt.figure(figsize=(12,5))
sns.histplot(df['bmi'],kde=True)
plt.title('BMI Distribution with KDE')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10,4))
sns.boxplot(x =df['bmi'])
plt.title('Box PLot of BMI')
plt.show()
```

```python
# remove outliers

Q1 = df['bmi'].quantile(0.25)
Q3 = df['bmi'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f'Lower Bound: {lower_bound}')
print(f'Upper Bound: {upper_bound}')
```

```python
df.drop_duplicates(inplace = True)
```

```python
# clipping ouliers
df_cleaned = df[(df['bmi'] >= lower_bound) & (df['bmi'] <= upper_bound)]
```

```python
plt.figure(figsize=(12,5))
sns.histplot(df_cleaned['bmi'],kde=True)
plt.title('BMI Distribution with KDE')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10,4))
sns.boxplot(x =df_cleaned['bmi'])
plt.title('Box PLot of BMI')
plt.show()
```

```python
df_cleaned
```

```python
# One hot encoding
df_cleaned = pd.get_dummies(df_cleaned,drop_first = True)
df_cleaned
```

```python
df_cleaned.corr() # corr -> creates metrix and find relation b/t each col
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_cleaned.corr(),annot = True)
```

```python
df_cleaned.columns
```

```python
x = df_cleaned[['age','bmi','smoker_yes']]
y = df_cleaned['charges']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 23)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
# testing

y_pred = model.predict(x_test)    # testing on test data
y_train_pred = model.predict(x_train)  # testing on train data
```

```python
print(f'Testing Result: {r2_score(y_test,y_pred)}')
print(f'Training Result: {r2_score(y_train,y_train_pred)}')
```

```python
print(f'Testing Result: {mean_squared_error(y_test,y_pred)}')
print(f'Training Result: {mean_squared_error(y_train,y_train_pred)}')
```

```python
import pickle

with open("Best_model.pkl",'wb') as f:
    pickle.dump(model,f)
```
